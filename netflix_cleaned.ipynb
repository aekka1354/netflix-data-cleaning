```python
import pandas as pd
```

```python
#import 'netflix_title.csv' file
df = pd.read_csv('netflix_titles.csv')
```

```python
#check heads
df.head()
```

```python
#check info
df.info()
```

```python
#describe info
df.info()
df.describe(include='all')
```

```python
#1) check for null values

df.isnull().sum()
```

```python
#2) drop missing title & type
df= df.dropna(subset=['title','type'])
```

```python
#3) reaplace null to not available in 'director' and 'cast' and 'country' columns
df['director']=df['director'].fillna("not available")
df['cast'] = df['cast'].fillna("not available")
df['country'] =df['country'].fillna("not available")
```

```python
#checking removed null values
df.isnull().sum() 
```

```python
#4) remove duplicate
df = df.drop_duplicates()
```

```python
#5) standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
#check
df.info() 
```

```python
#6) convert date formate
df["date_added"]=pd.to_datetime(df['date_added'],errors='coerce')
```

```python
#chekcing dtype of date_added is changed or not
df.info()
```

```python
#7) clean 'type' column
df['type'] = df['type'].str.strip().str.title()
```
